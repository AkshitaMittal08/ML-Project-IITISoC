```python
#Defining a function to extract relevant features of the images using pre-trained model Inception V3
```

```python

def image_feature(path):
    model= InceptionV3(weights='imagenet',include_top= False)
    features=[];
    img_name=[];
    for folder in os.listdir(path):
        sub_path=path+"/"+folder
        for img in os.listdir(sub_path):
            image_path=sub_path+"/"+img
            img_arr=cv2.imread(image_path)
            img_arr=cv2.resize(img_arr,(150,150))
            img_arr=np.expand_dims(img_arr,axis=0)
            img_arr=preprocess_input(img_arr)
            feat=model.predict(img_arr)
            feat=feat.flatten()
            features.append(feat)
            img_name.append(img)
    return features,img_name
        
        
```

```python
path= '../input/datasetiitisocac'
img_features,img_names=image_feature(path)
```

```python
import pandas as pd
```

```python
#Creating a pandas dataframe to store columns of the two lists
```

```python
make_list=pd.DataFrame(list(zip(img_names,img_features)),columns=['Img_Names','Img_features'])
```

```python
#Shuffling the dataset so that the model does not learn anything from the order of the images
```

```python
shuffled= make_list.sample(frac=1)
print(shuffled.head())
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(shuffled_list)
shuffled_list_sc = sc.transform(shuffled_list)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
```

```python
shuffled_list_pca=pca.fit_transform(shuffled_list_sc)
```

```python
#Creating Clusters
#Implementing kMeans clustering algo
k = 10
clusters= KMeans(n_clusters=10, random_state=0, init='k-means++', max_iter=300, n_init=10)

clusters.fit_predict(shuffled_list_pca)
```

```python
img_names_shuffled_list= list(shuffled['Img_Names'])
image_cluster = pd.DataFrame(img_names_shuffled_list,columns=['image'])
image_cluster["clusterid"] = clusters.labels_
image_cluster 
```
